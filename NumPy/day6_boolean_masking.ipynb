# Day 6 – Boolean Masking and Filtering

## Objective
Learn how to filter data, apply conditions, replace values, and handle missing values using NumPy.

**Name:** Ponmalar

**Day:** 6


```python
import numpy as np
marks = np.array([45, 67, 89, 34, 92, 76, 58, 99, 81, 39])

print(marks)
```

```python
print(marks > 70)
```

```python
print(marks[marks > 70])
```

```python
print(marks[marks < 50])
```

```python
print(marks[(marks > 60) & (marks < 90)])
```

```python
print(marks[(marks < 40) | (marks > 90)])
```

```python
marks[marks < 40] = 40

print(marks)
```

```python
data = np.array([10,20,np.nan,40,50,np.nan])

print(data)
```

```python
print(np.isnan(data))
```

```python
data[np.isnan(data)] = 0

print(data)
```

# 📊 Mini Project – Employee Performance Analysis

## Objective
Analyze employee performance scores using NumPy. Filter employees based on performance, identify top performers, update low scores, and handle missing values.

### Concepts Used
- NumPy Arrays
- Boolean Masking
- Filtering
- Multiple Conditions
- Missing Values (NaN)
- Statistical Functions

### Problem Statement
A company wants to analyze employee performance scores. HR needs to identify high-performing employees, employees needing training, calculate average performance, and replace missing scores before generating a report.

```python
import numpy as np
scores = np.array([78, 92, 55, 81, 45, 99, 67, 88, 39, 74])

print("Employee Scores")
print(scores)
```

```python
print("Highest Score:", np.max(scores))
print("Lowest Score:", np.min(scores))
print("Average Score:", np.mean(scores))
print("Total Score:", np.sum(scores))
```

```python
print("Employees scoring above 80")
print(scores[scores > 80])
```

```python
print("Employees scoring below 50")
print(scores[scores < 50])
```

```python
print(scores[(scores >= 60) & (scores <= 90)])
```

```python
scores[scores < 50] = 50

print(scores)
```

```python
attendance = np.array([98, np.nan, 87, 91, np.nan, 95])

print(attendance)
```

```python
print(np.isnan(attendance))
```

```python
attendance[np.isnan(attendance)] = 0

print(attendance)
```

## 📋 Final Performance Report

```python
print("===== PERFORMANCE REPORT =====")

print("Highest Score :", np.max(scores))
print("Lowest Score :", np.min(scores))
print("Average Score :", np.mean(scores))
print("Employees Above 80 :", scores[scores > 80])
print("Employees Below 60 :", scores[scores < 60])
```
