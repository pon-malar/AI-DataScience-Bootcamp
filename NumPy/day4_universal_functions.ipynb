# 📊 Mini Project – Student Marks Analysis

## 🎯 Objective
Analyze a dataset of student marks using NumPy and generate useful insights using statistical functions, indexing, broadcasting, and filtering.

### Concepts Used
- NumPy Arrays
- Indexing and Slicing
- Broadcasting
- Maximum and Minimum
- Mean and Median
- Sum
- Sorting
- Unique Values
- Argmax
- Boolean Indexing
- np.clip()

### Problem Statement
A teacher wants to analyze the marks of students in a class. The goal is to identify the highest and lowest marks, calculate the average, find top-performing students, identify students who need additional support, and generate a final report after awarding grace marks.

### Expected Outcome
By the end of this project, you will be able to:
- Analyze student marks using NumPy.
- Perform statistical calculations.
- Filter data based on conditions.
- Apply broadcasting to update marks.
- Generate a simple performance report.

```python
import numpy as np

marks = np.array([78, 92, 65, 88, 95, 70, 88, 81, 76, 90])

print("Student Marks")
print(marks)
```

## 📈 Task 1: Basic Statistical Analysis

Calculate the highest mark, lowest mark, total marks, average mark, and median of the dataset.

```python
print("Highest Mark:", np.max(marks))
print("Lowest Mark:", np.min(marks))
print("Total Marks:", np.sum(marks))
print("Average Mark:", np.mean(marks))
print("Median Mark:", np.median(marks))
```

## 🏆 Task 2: Identify High Performing Students

Find all students who scored more than 85 marks.

```python
print("Students scoring above 85:")
print(marks[marks > 85])
```

## 📚 Task 3: Identify Students Needing Improvement

Find students who scored below 70 marks.


```python
print("Students scoring below 70:")
print(marks[marks < 70])
```

## 📊 Task 4: Sort and Remove Duplicate Marks

Sort the marks in ascending order and display only unique marks.

```python
print("Sorted Marks:")
print(np.sort(marks))

print("Unique Marks:")
print(np.unique(marks))
```

## 🥇 Task 5: Find the Top Scorer

Identify the position (index) of the student with the highest mark.

```python
highest_index = np.argmax(marks)

print("Top Scorer Index:", highest_index)
print("Top Score:", marks[highest_index])
```

## ➕ Task 6: Award Grace Marks

Add 5 grace marks to every student using NumPy broadcasting.

```python
updated_marks = marks + 5

print(updated_marks)
```

## 🎯 Task 7: Limit Maximum Marks

Ensure that no student's marks exceed 100 using the `np.clip()` function.

```python
final_marks = np.clip(updated_marks, 0, 100)

print(final_marks)
```

## 📋 Final Performance Report

Generate a summary report after applying grace marks.

```python
print("===== FINAL REPORT =====")
print("Highest Mark :", np.max(final_marks))
print("Lowest Mark  :", np.min(final_marks))
print("Average Mark :", np.mean(final_marks))
print("Students Scoring Above 90 :", final_marks[final_marks >= 90])
```

# ✅ Conclusion

In this mini project, student marks were analyzed using NumPy. Statistical measures such as maximum, minimum, mean, median, and total were calculated. High-performing students and students needing improvement were identified. Broadcasting was used to award grace marks, and `np.clip()` ensured that marks remained within the valid range. This project demonstrates how NumPy can efficiently analyze and manipulate numerical datasets.
