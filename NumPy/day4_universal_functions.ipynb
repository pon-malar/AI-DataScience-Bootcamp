```python
import numpy as np

marks = np.array([78, 92, 65, 88, 95, 70, 88])

print("Marks:")
print(marks)
```

```python
print("Maximum:", np.max(marks))
print("Minimum:", np.min(marks))
```

```python
print("Sum:", np.sum(marks))
print("Average:", np.mean(marks))
```

```python
print("Median:", np.median(marks))
```

```python
print("Standard Deviation:", np.std(marks))
print("Variance:", np.var(marks))
```

```python
print("Index of Maximum:", np.argmax(marks))
print("Index of Minimum:", np.argmin(marks))
```

```python
print(np.sort(marks))
```

```python
print(np.unique(marks))
```

```python
print(np.where(marks > 80))
```

```python
print(np.clip(marks,70,90))
```

```python
sales = np.array([25000,32000,28000,35000,30000])

print("Sales:", sales)
print("Highest Sales:", np.max(sales))
print("Lowest Sales:", np.min(sales))
print("Average Sales:", np.mean(sales))
print("Total Sales:", np.sum(sales))
```
