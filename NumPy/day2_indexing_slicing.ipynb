```python
import numpy as np

numbers = np.array([10, 20, 30, 40, 50])

print(numbers[0])
print(numbers[2])
print(numbers[-1])
print(numbers[-2])
```

```python
# Slicing

print(numbers[1:4])
print(numbers[:3])
print(numbers[2:])
print(numbers[::-1])
```

```python
numbers[0] = 100
numbers[2] = 300

print(numbers)
```

```python
import numpy as np

matrix = np.array([
    [10, 20, 30],
    [40, 50, 60],
    [70, 80, 90]
])

print(matrix)

print("First row:", matrix[0])
print("Second row:", matrix[1])

print("First element:", matrix[0,0])
print("Center element:", matrix[1,1])
print("Last element:", matrix[2,2])
```

```python
print(matrix[:, 0])

print(matrix[1, :])

print(matrix[0:2, 1:3])

print(matrix[:, 1:])
```

```python
matrix[0,0] = 100
matrix[2,1] = 800

print(matrix)
```

```python
import numpy as np

marks = np.array([75, 92, 81, 95, 67, 88])

print("Original Marks:")
print(marks)

print("\nMarks greater than 85:")
print(marks[marks > 85])
```

```python
numbers = np.array([10,20,30,40,50,60])

print(numbers[[0,2,5]])
```
